```python
# get libraries
from urllib.request import urlopen
import urllib.request
import requests

import random
import time

import re
import os
```

```python
def get_decree_id(url):
    # Extract the decree code using regex
    match = re.search(r"_codeArrete=(\d+)", url)
    if match:
        code = match.group(1)
        return f"{code}"
    else:
        return None
```

```python
def generate_decree_filename(url, base_name='arrete_', extension='csv'):
    # Extract the decree number using regex
    return f"{base_name}{code}.{extension}"   
```

```python
def generate_urls_list(url, start=1, end=859):
    # return  list of urls containing decrees from start to end
    
    base_url = url.split(get_decree_id(url))[0]
    tail_url = url.split(get_decree_id(url))[1]

    # Initialize an empty list to store the URLs
    url_list = []

    # Generate URLs with codeArrete values from 1 to 859
    for code in range(1, 860):
        url = f"{base_url}{code:06d}{tail_url}"
        url_list.append(url)

    return url_list
```

```python
def download_decrees(urls, folder_dest=decrees_folder_name, base_name=decree_filename_base, extension_name=decree_extension_name, n=start, m=stop, enable_random_wait=True):
    # Download decrees n to m from urls list.
    
    #create folder_dest if needed
    if not os.path.exists(folder_dest):
        os.makedirs(folder_dest)
    for url in urls[n-1:m]:    
        local_filename = f"{base_name}{get_decree_id(url)}.{extension_name}"
        print(local_filename)
        path = os.path.join(folder_dest, local_filename)
        print(url)
        print(path)

        # Download the CSV file
        urllib.request.urlretrieve(url, path)
    

        secondstowait = random.randint(10, 2000)/1000


        # Sleep for the total random time
        if enable_random_wait:
            time.sleep(secondstowait)
            print(secondstowait)   
    
```
